```python
def removeStopword(text):
    word_tokenize=RegexpTokenizer(r'\w+')
    stop_words = set(stopwords.words('english'))
    stop_words.update(string.punctuation)
    word_tokens = word_tokenize.tokenize(text) 
    filtered_sentence = [word for word in word_tokens if not word in stop_words]  
    return filtered_sentence
```

```python

questions_df["tokens"] = questions_df["body"].apply(lambda x : removeStopword(x))
test["tokens"] = test["body"].apply(lambda x : removeStopword(x))
```

```python
y_train = questions_df.type.map({'summary':1,'factoid':2,'list':3,'yesno':4}).tolist()
y_test = test.type.map({'summary':1,'factoid':2,'list':3,'yesno':4}).tolist()
```

```python
num_classes = np.max(y_train)+1
max_words=798
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=max_words)
```

```python
# x_train=tokenizer.sequences_to_matrix(x_train,mode='binary')
# x_test=tokenizer.sequences_to_matrix(x_train,mode='binary')
x_train = tokenizer.sequences_to_matrix(x_train, mode='binary')
x_test = tokenizer.sequences_to_matrix(x_test, mode='binary')
```

```python
y_train
```

```python
y_train = to_categorical(y_train,num_classes)
y_test = to_categorical(y_test,num_classes)
```

```python
y_train
```

```python
model = Sequential()# Instantiate sequential model
model.add(Dense(512,activation="relu",input_shape=(max_words,)))


model.add(Dense(256,activation="relu"))


model.add(Dense(num_classes, activation="softmax")) # Add third layer
```

```python
opt = optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["accuracy"])
```

```python
batch_size = 32
model.fit(x_train,y_train,batch_size=batch_size,
         epochs=10,validation_data=(x_test,y_test))
score = model.evaluate(x_test,y_test,verbose=1)
```
